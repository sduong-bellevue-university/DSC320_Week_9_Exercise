```python
plt.figure(figsize=(10, 6))                                                           # create new figure with specified size

boxplot = plt.boxplot(qb_df['td'], vert=False, patch_artist=True,                     # create a horizontal boxplot of the touchdowns (td) column                
                      medianprops=dict(color='red', linewidth=2),                     # set the color of the median line to 'red' and its linewidth to 2
                      flierprops=dict(marker='o', markerfacecolor='red'))             # set the marker for outliers to a circle and the marker face color to red

plt.title("Boxplot of Touchdowns for NFL Quarterbacks (2019 Regular Season)")         # create the plot title
plt.xlabel("Number of Touchdowns")                                                    # create the x-axis label passing yards
plt.yticks([])                                                                        # remove y-axis label since it's not needed for a single boxplot
plt.grid(axis="x", alpha=0.3, linestyle='--')                                         # display grid lines along x axis and set transparency
plt.tight_layout()                                                                    # adjust parameters to ensure labels and titles fit within the figure area
plt.show()                                                                            # display the graph
```

```python
Q1, Q3 = qb_df['td'].quantile([0.25, 0.75])                                 # calculate the 25th percentile (Q1) and 75th percentile (Q3) using quantile() method
IQR = Q3 - Q1                                                               # calculate interquartile range (IQR) for the range of the middle 50% of the data
lower_bound = Q1 - 1.5 * IQR                                                # define lower bound (1.5 * IQR below Q1)          
upper_bound = Q3 + 1.5 * IQR                                                # define upper bound (1.5 * IQR above Q3)
outliers = qb_df[(qb_df['td'] < lower_bound) | (qb_df['td'] > upper_bound)] # filter outliers where touchdowns are below lower bound or above upper bound

print(f"Lower Bound: {lower_bound:.2f}")                             # print lower bound
print(f"Upper Bound: {upper_bound:.2f}")                             # print upper bound 


print(f"\nNumber of Outliers: {len(outliers)}")                      # print number of outliers
if len(outliers) > 0:
    # display the outliers with the quarterback's name, team, and touchdown information
    for idx, row in outliers.iterrows():
        print(f"{row['name']} ({row['team']}): {row['td']:.0f} touchdowns")
else:
    print("No outliers detected in the touchdown data.")             # display message indicating no outliers detected
```

### 2. Calculating Probabilities from Data

#### (a) Based off of this data, what is the probability a college student has brown hair?

```python
# import libraries
import numpy as np # library for working with matrices and arrays
import pandas as pd # library for working with data frames
import matplotlib.pyplot as plt # library for visualizations
```

```python
# import data from the survey_data.csv file
survey_df = pd.read_csv("survey_data.csv")
```

```python
# display the first few rows to verify the data loaded correctly
survey_df.head()
```

```python
# count the frequency of each hair color in the dataset
hair_color_counts = survey_df["hair_color"].value_counts()
print("Hair color distribution:")
print(hair_color_counts)

# determine the total number of students surveyed using shape of the dataset
total_students = len(survey_df)
print(f"Total number of students surveyed: {total_students}")
```

```python
# calculate the number of students with brown hair
#brown_hair_count = survey_df["hair_color"].str.lower().str.strip().eq("brown").sum()
brown_hair_count = survey_df["hair_color"].eq("brown").sum()
print(f"Number of colege students with brown hair: {brown_hair_count}")

# calculate the probability of a student having brown hair
probability_brown_hair = brown_hair_count / total_students

# display the probability as a decimal and percentage
print(f"Probability of college students with brown hair (decimal): {probability_brown_hair:.4f}")
print(f"Probability of college students with brown hair (percentage): {probability_brown_hair * 100:.2f}%")
```

#### (b) Based off of this data, what is the probability a college student has blue eyes?

```python
# calculate the number of students with blue eyes
blue_eyes_count = survey_df["eye_color"].eq("blue").sum()
print(f"Number of college students with blue eyes: {blue_eyes_count}")

# calculate the probability of a student having blue eyes
probability_blue_eyes = blue_eyes_count / total_students

# display the probability as a decimal and percentage
print(f"Probability of college students with blue eyes (decimal): {probability_blue_eyes:.4f}")
print(f"Probability of college students with blue eyes (percentage): {probability_blue_eyes * 100:.2f}%")
```

#### (c) Based off of this data, what is the probability a college student has blue eyes given that they have brown hair?

```python
# filter the dataset to include only college students with brown hair
brown_hair_df = survey_df[survey_df["hair_color"] == "brown"]
print(f"Number of college students with brown hair: {len(brown_hair_df)}")

# count students with blue eyes amongst those with brown hair
blue_eyes_brown_hair_count = (brown_hair_df["eye_color"] == "blue").sum()
print(f"Number of college students with brown hair and blue eyes: {blue_eyes_brown_hair_count}")

# calculate conditional probability P(blue eyes | brown hair)
probability_blue_given_brown = blue_eyes_brown_hair_count / len(brown_hair_df)

# display the conditional probability P(blue eyes | brown hair) as a decimal and percentage
print(f"Conditional probability of college students with blue eyes given brown hair (decimal): {probability_blue_given_brown:.4f}")
print(f"Conditional probability of college students with blue eyes given brown hair (percentage): {probability_blue_given_brown * 100:.2f}%")
```

#### (d) Based off of this data, what is the probability a college student has brown hair given that they have blue eyes?
